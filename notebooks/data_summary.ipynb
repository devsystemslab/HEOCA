## plot for main figure

```python
all_samples = pd.read_csv(f"{project_dir}/all_samples_sheets.txt", sep="\t")
all_samples.head()
```

```python
# adata_obs.groupby('sample_id').count()
```

```python
all_samples = pd.merge(all_samples, 
        pd.DataFrame(adata_obs.groupby('sample_id').count()['leiden']).reset_index(),
         on='sample_id', how='left')
```

```python
all_samples = all_samples.rename(columns={'leiden':'cell_number'})
```

```python
new_protocol=[]
new_time=[]
for i,j,k in zip(all_samples.derive, all_samples.time, all_samples.protocol):
    if i=='ASC':
        new_protocol.append('ASC')
        new_time.append(130)
    if i=='FSC':
        if str(j)=='nan':
            new_protocol.append('FSC_late')
            new_time.append(120)
        if j=='early':
            new_protocol.append('FSC_early')
            new_time.append(110)
        if j=='late':
            new_protocol.append('FSC_late')
            new_time.append(120)
    if i=='IPS':
        if k=='transplant':
            new_protocol.append('IPS_trans')
            if str(j)=='nan':
                new_time.append(1)
            else:
                new_time.append(int(j[:-3]))
        else:
            new_protocol.append('IPS')
            if str(j)=='nan':
                new_time.append(1)
            else:
                new_time.append(int(j[:-3]))            
```

```python
all_samples['new_protocol']=new_protocol
all_samples['new_time']=new_time
all_samples['new_protocol_time'] = all_samples['new_protocol'] + "_" + all_samples['new_time'].astype(str) 
all_samples['tissue2'] = all_samples['tissue'] + "_" + all_samples['tech_detail'].astype(str) 
```

```python
all_samples.groupby(['tissue','tissue2', 'new_protocol', 'publication',
                                    'new_time', 'tech_detail']).sum()
```

```python
all_samples2 = all_samples.groupby(['tissue','tissue2', 'derive', 'new_protocol', 'publication',
                                    'new_time', 'tech_detail']).count()['sample'].reset_index()

all_samples2['cell_number'] = all_samples.groupby(['tissue','tissue2', 'new_protocol', 'publication',
                                    'new_time', 'tech_detail']).sum()['cell_number'].tolist()

all_samples2 = all_samples2.rename(columns={'sample':'sample_number'})
all_samples2['point_sizes']=[i+10 if i>1 else 1 for i in all_samples2.sample_number]
all_samples2['sizes']=[i if i>0 else '' for i in all_samples2.sample_number]
all_samples2.head()
```

```python
rcParams['figure.dpi'] = 100

fig, axs = plt.subplots(1, 2, figsize=(8, 3), constrained_layout=True)
sns.set_style("white", {'axes.grid' : False})


forplot2 = all_samples2.groupby(['tech_detail']).sum().reset_index()
forplot2 = forplot2.sort_values('cell_number', ascending=False)
forplot2['log10_cell_number'] = round(np.log10(forplot2['cell_number']),2)

sns.barplot(data=forplot2, y="tech_detail", x="sample_number",ax=axs[0])
# axs[0,1].set_xscale("log")
# axs[0,1].set(xscale="log")
axs[0].set_ylabel("scRNA-seq methods")


forplot5 = all_samples2.groupby(['tissue','derive']).sum().reset_index()
forplot5 = forplot5.sort_values('cell_number', ascending=False)
forplot5['log10_cell_number'] = round(np.log10(forplot5['cell_number']),2)

sns.barplot(data=forplot5, y="tissue", x="sample_number",hue='derive', 
            palette=plate_level_all, width=1, ax=axs[1])

# axs[1,1].set_xscale("log")
# axs[1,1].set(xscale="log")
axs[1].set_ylabel("scRNA-seq tissues")
# axs[1].get_legend().remove()
# axs[1,1].set_xlim(0, 100000)

fig.savefig(f"{project_dir}/figures/heoa/sample_info1.pdf", bbox_inches="tight")
```
