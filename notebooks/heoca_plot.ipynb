## function

```python
def get_bulk_exp(adata, bulk_labels, layer='var'):
    if layer=='raw':
        res = pd.DataFrame(columns=adata.raw.var_names, index=adata.obs[bulk_labels].cat.categories)
    else:
        res = pd.DataFrame(columns=adata.var_names, index=adata.obs[bulk_labels].cat.categories)                                                                                                 
    
    for clust in adata.obs[bulk_labels].cat.categories: 
        if layer=='raw':
            res.loc[clust] = adata[adata.obs[bulk_labels].isin([clust]),:].raw.X.mean(0)
        else:
            res.loc[clust] = adata[adata.obs[bulk_labels].isin([clust]),:].X.mean(0)

    res.index=adata.obs[bulk_labels].cat.categories

    return res.T.astype(float)
```

### plot sankey plot

```python
min_ct_perc = 0.02

y_lev1 = adata.obs.level_1_late
lev1_percs = {ct: n / len(y_lev1) * 100 for ct, n in Counter(y_lev1).items()}
lev1_ct_to_keep = [ct for ct, perc in lev1_percs.items() if perc > min_ct_perc]

# get level 1 anns, set "None" in level 2 compartment specific,
# remove cell types that make up less than min_ct_perc of cells plotted
y_lev2 = adata.obs.level_1_late.cat.remove_unused_categories()
y_lev2 = [
    f"{ct} ({lev1ann})" if ct == "None" else ct
    for ct, lev1ann in zip(y_lev2, adata.obs.level_1_late)
]
lev2_percs = {ct: n / len(y_lev2) * 100 for ct, n in Counter(y_lev2).items()}
lev2_ct_to_keep = [ct for ct, perc in lev2_percs.items() if perc > min_ct_perc]


fig, ax = plt.subplots(figsize=(6, 10))

sankey.sankey(
    x=[lev1
       for lev1, lev2 in zip(y_lev1, list(y_lev2))
       if lev1 in lev1_ct_to_keep and lev2 in lev2_ct_to_keep],
    y=[lev2
        for lev1, lev2 in zip(y_lev1, list(y_lev2))
        if lev1 in lev1_ct_to_keep and lev2 in lev2_ct_to_keep],
    title="Hierarchical \nannotation",
    title_left="Level 1",
    title_right="Level 2",
    ax=ax,
    fontsize="x-small",
    colorside='right',
    colors=plate_level_all,
    alpha=1,
)
plt.show()
plt.close()

```

#### load markers

```python
topmarkers = pd.read_csv(f'{project_dir}/results/{results_dir}/gut_scpoli_integration_100k/topmarkers.txt', 
                         sep='\t')
topmarkers = topmarkers.drop(columns=['rank'])
topmarkers = topmarkers.to_dict()

topmarkers2={}
for i in topmarkers:
    topmarkers2[i] = list(topmarkers[i].values())
```

```python
bulk_exp = get_bulk_exp(adata100k, 'level_2_late', layer='raw')
```

```python
alltissues = ['intestine', 'liver', 'lung', 'pancreas', 'prostate', 
              'salivarygland', 'stomach', 'thyroid']

show_markers=[]
for tissue in alltissues:
    marker_genes = read_yaml(f"{project_dir}/markers/{tissue}_markers.yaml")

    for i in marker_genes:
        for j in get_markers(marker_genes):
            show_markers+=get_markers(marker_genes)[j]


        if 'subtypes' in marker_genes[i]:
            for k in get_markers(marker_genes[i]['subtypes']):
                show_markers+=get_markers(marker_genes[i]['subtypes'])[k]
```

```python

topmarkers3=[]
for i in topmarkers2:
    topmarkers3+=topmarkers2[i]
    
forplot = bulk_exp.loc[topmarkers3,:].drop_duplicates().T

forplot -= forplot.min(0)
forplot = (forplot / forplot.max(0)).fillna(0)

Z = linkage(forplot.to_numpy())
cell_order = forplot.iloc[leaves_list(Z),:].index[::-1]

topmarkers4=[]
for i in cell_order:
    topmarkers4+=topmarkers2[i]

gene_order=[]
for i in topmarkers4:
    if i not in gene_order:
        gene_order.append(i)

rmgenes=[]

fig, ax = plt.subplots(figsize=(20, 8))
sns.heatmap(forplot.loc[cell_order, gene_order], cmap='Blues', ax=ax, 
            xticklabels=True, yticklabels=True)

ax.set_xticklabels([i if (i in show_markers) and (i not in rmgenes) else '' for i in gene_order]);
ax.tick_params(bottom=False)
```

```python
fig, ax = plt.subplots(figsize=(2, 8))
dn = dendrogram(Z, no_labels=True, orientation='right', ax=ax)
ax.set_xticklabels([]);
ax.set_yticklabels([]);
ax.axis('off')

ax.tick_params(bottom=False, left=False)
```
